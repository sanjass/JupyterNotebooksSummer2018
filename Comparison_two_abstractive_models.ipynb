## Comparison of abstractive summaries using two models and variations

#### Sanja Simonovikj - Shell internship, summer 2018


Here we will qualitatively compare the results of two models presented in the papers:

#### 1. [Fast Abstractive Summarization with Reinforce-Selected Sentence Rewriting](https://arxiv.org/pdf/1805.11080.pdf)
- referenced as fast_abs_rl

#### 2. [Get To The Point: Summarization with Pointer-Generator Networks](https://arxiv.org/pdf/1704.04368.pdf)
- referenced as pointer-generator



For each of the two models, we will try applying [TextRank](https://web.eecs.umich.edu/~mihalcea/papers/mihalcea.emnlp04.pdf) on the original document, before feeding the text into the networks, and see if this improves the final abstractive summary. 

Note: When running TextRank we set the limit of the produced summary to be ~150 words.

**Original text**
> http://www.energyintel.com/pages/SearchArticleView.aspx?DocID=993888&isSearchResult=true
Eni Agrees Lower Costs for Indonesian Gas Field
2018/03/29
Eni has reduced expected costs for its Merakes gas field development offshore Indonesia by around 14% to \$1.03 billion after its original development plan was rejected by the government as too expensive, an official with the country's energy ministry has told International Oil Daily.
The Italian major's plan for Merakes was submitted to the energy ministry for approval earlier this month, with an estimated budget of around $1.2 billion.
But with the government on the hook to reimburse Eni for development costs, the ministry asked Eni to revise its plans and come back with a lower figure (IOD Mar.23'18).
"The energy ministry has reviewed the cost of the Merakes project and Eni has agreed with the results.
Eni will soon propose a new development plan," the official, who declined to be named, told IOD on Wednesday.
While it remains unclear how the cost savings were found or whether some compromise was reached, the swift progress toward a new development plan underlines Jakarta's determination to keep Merakes on track to help it offset falling output from other fields, and Eni's desire to make progress with one of six projects penciled in for a final investment decision (FID) this year as part of broader push to develop its gas and power portfolio (EIF Mar.21'18).
The ministry now expects Eni to take FID on Merakes by May, assuming there are no other major obstacles, the official said.
"The energy ministry wants Merakes to come on stream in 2019.
That target is still in line with the development plan," the official said.
Eni spokesman Filippo Cotalini declined to comment on the development plan but said the company “will communicate a possible update on Merakes in due time.”
Merakes, located in the Sepinggan Block off East Kalimantan, is expected to produce about 150 million cubic feet per day of gas, which will be supplied to the Bontang LNG plant, and is estimated to hold around 2 trillion cubic feet of reserves.
Eni has an 85% operating interest while state-run Pertamina holds 15% (IOD Jun.26'17).
Eni's development plan envisages a new platform linked to a floating production unit and other pipeline facilities, according to the energy ministry.
The energy ministry's newly appointed director general of oil and gas, Djoko Siswanto, told reporters on Wednesday that he would ensure that the Merakes project proceeds smoothly in order to meet the planned start-up date.
Changes to the country's production-sharing contract model have deterred investors, who say Indonesia's fiscal terms are becoming less attractive and more stringent, exacerbating the problems caused by lower oil and gas prices.
Several companies including Exxon Mobil, ConocoPhillips and Hess have divested or walked away from assets in recent years, and there is still uncertainty over the timing of FIDs for several major upstream projects, including Inpex's Abadi project, and Chevron's Gendalo-Gehem field (IOD Mar.6'18).
Muklis Ali, Jakarta

Oil Slips, Weighed by Saudi Output, Trade Spat
2018/07/06
U.S. crude inventories rise unexpectedly; U.S. implements tariffs on Chinese goods on Friday.
  Oil slipped below \$77 a barrel on Friday, under pressure from higher Saudi production and trade tensions between the United States and China, despite support from oil supply disruptions.
  Top exporter Saudi Arabia told OPEC it raised oil output by almost 500,000 barrels per day last month, OPEC sources said, a sign Riyadh wants to make up for shortages elsewhere and dampen prices.
  Brent crude, the global benchmark, was down 76 cents at \$76.63 a barrel by 1148 GMT. U.S. crude slipped 60 cents to \$72.34.

  A U.S. government report also weighed on prices this week, showing crude stockpiles rose by 1.3 million barrels, while analysts had forecast a decline.
  The potential trade war between the United States and China comes amid a tight oil market.
  Oil output cuts by the Organization of Petroleum Exporting Countries and allies including Russia since January 2017 have reduced a glut of crude.
  Involuntary drops in supply in Venezuela, Angola and Libya have made the cutbacks even bigger, although OPEC has now started to ease those curbs with Saudi Arabia pumping more.
  Even so, renewed U.S. sanctions on Iran against its oil exports look set to tighten supply further.

Exxon Makes Fifth Guyana Oil Find
2017/10/06
Exxon Mobil has announced yet another find offshore Guyana at the Turbot-1 well about 30 miles southeast of the Liza prospect (OD Jul.26'17).
The US major said it began drilling the well on Aug. 14 and encountered a 75-foot reservoir of high-quality oil-bearing sandstone.
The well was drilled to a total depth of 18,445 feet in 5,912 feet of water.
The well was completed on Sep. 29.
The successful well adds to discoveries to date at the Liza, Payara, Snoek and Liza Deep prospects.
However, despite the string of promising finds, Guyana is still working to lay the foundations of its oil and gas sector.
After the state took public comment over the summer on a draft petroleum law, the bill itself is still sitting in committee in the country’s parliament, awaiting legislators’ next moves.
In speaking to local press outfit Oil Now, Natural Resource Minister Raphael Trotman said Guyana had yet to even figure out who will market its oil, even though production is slated to commence in 2020.
Estimates from Oil Now suggest the government take on the very first day of production in 2020 would be just under \$900,000 at current barrel prices.
The calculation is based on projected output of 120,000 b/d, as well as a negotiated 2% royalty and 50% "profit oil" share.
Oil fever has certainly gripped tiny Guyana, which has about 750,000 residents according to World Bank data.
The government even had to recently warn citizens about fraudulent technical programs that have popped up, offering training for energy sector jobs.
Citizens hungry for work have been pushing the government to press Exxon on local content provisions, even though officials have already publicly acknowledged that the potential for job creation may be limited (OD Jun.12'17).
Meanwhile, Exxon as of late has made other big moves to add South American assets to a portfolio that for some time was woefully short on them.
Most notably, the supermajor spent big alongside state-owned oil producer Petrobras at a recent auction, snagging six blocks offshore Brazil that may have opened a “back door” to the prolific presalt formation (OD Oct.4'17).
Exxon said after the Turbot-1 well is completed offshore Guyana, the Stena Carron drillship will move to the Ranger prospect.
It said an additional well at Turbot is being planned for 2018.
Anthony Venezia, New York
