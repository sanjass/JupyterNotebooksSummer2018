### Sentiment analysis
We will try two approaches for getting sentiment analysis score:
- [TextBlob](https://github.com/sloria/TextBlob)
> The sentiment property returns a named tuple of the form Sentiment(polarity, subjectivity). The polarity score is a float within the range [-1.0, 1.0]. The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.
- [Standford CoreNLP Sentiment Analysis](https://nlp.stanford.edu/sentiment/)
> The average sentiment of a sentence is between Positive (3), Neutral (2) and Negative (1), the range is from VeryNegative (0) to VeryPositive (4) which appear to be quite rare.
Note: [python usage here](https://www.quora.com/Could-anyone-give-me-an-example-of-using-Stanford-CoreNLP-sentiment-analysis-with-Python)
